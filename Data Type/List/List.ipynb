# List

## 列表的定义

```python
MyList1 = list(range(1,10))
MyList2 = list((1,2,"foo","bar"))

print(MyList1, MyList2)
```

## 列表的操作

### 增加

```python
List = [1,2,3,4]
List1,List2,List3 = [List.copy(), List.copy(), List.copy()]
List1.append(5)
List2.insert(4,5)
print(List1, List2)
```

```python
[1,2,3] + [4,5,6,7]

## equivalent to this:
x = [1,2,3]
x.extend([4,5,6,7])
x
```

### 删除

```python
lst = [1,2,3,4,5,6]
lst.pop(0)
lst.remove(6)
lst
```

## 列表的排序

```python
[1,4,3,5].sort()
```

## 列表的切片

```python
lst = [1,2,3,4,5,6,7,8,9]
lst[3:5]
lst[-1]
```
